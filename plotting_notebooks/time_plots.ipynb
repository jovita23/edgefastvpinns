```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'EdgeVPINNs\time.csv')
```

```python
#plot line graph rank vs mean_time_per_epoch for each decomposition_type so 3 different plots
decomposition_types = df.decomposition_type.unique()
for decomposition_type in decomposition_types:
    df_decomposition_type = df[df['decomposition_type'] == decomposition_type]
    df_decomposition_type = df_decomposition_type.sort_values(by=['rank'])
    plt.plot(df_decomposition_type['rank'], df_decomposition_type['mean_time_per_epoch'])
    plt.xlabel('Rank')
    plt.ylabel('Mean Time Per Epoch')
    plt.title('Mean Time Per Epoch vs Rank for ' + decomposition_type)
    plt.show()
```
