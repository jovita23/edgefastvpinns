```python
def parse_rank_list(rank_list):
    try:
        # Remove surrounding quotes and spaces, then split and convert to integers
        return [int(x) for x in rank_list.strip('[]').split(',')]
    except Exception as e:
        print(f"Error parsing rank_list: {rank_list}. Error: {e}")
        return []
```
