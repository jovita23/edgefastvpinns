```python
import numpy as np
import matplotlib.pyplot as plt

# Generate grid points
x = np.linspace(0, 1, 100)
y = np.linspace(0, 1, 100)
X, Y = np.meshgrid(x, y)
test_points = np.vstack([X.flatten(), Y.flatten()]).T

# Define the exact solution function
def f(x):
    return -1.0 * np.sin(4.0 * np.pi * x[:, 0]) * np.sin(4.0 * np.pi * x[:, 1])

# Create subplots (1 row, 3 columns)
fig, axes = plt.subplots(2, 3, figsize=(15, 8),dpi = 300)

plt.rcParams['xtick.labelsize'] = 18
plt.rcParams['axes.titlesize'] = 18
plt.rcParams['axes.labelsize'] = 18
plt.rcParams['legend.fontsize'] = 18
plt.rcParams['ytick.labelsize'] = 18
plt.rcParams['xtick.labelsize'] = 18

# Plot the exact solution on ax1
y_exact = f(test_points)
Y_Exact_Matrix = y_exact.reshape(100, 100)
contour1 = axes[0,0].contourf(X, Y, Y_Exact_Matrix, levels=100, cmap='jet')
axes[0,0].set_title('(a)', fontsize=18)
axes[0,0].set_xlabel('x', fontsize=18)
axes[0,0].set_ylabel('y', fontsize=18)
axes[0,0].set_xticks([0, 0.5, 1])
axes[0,0].set_yticks([0, 0.5, 1])
axes[0,0].set_aspect('equal')
axes[0,0].set_xticklabels(['0', '0.5', '1'])
plt.colorbar(contour1, ax=axes[0,0])  # Add colorbar to ax1

# Plot the predicted solution on ax2
y_pred = np.loadtxt("/home/jovita/Projects/EdgeVPINNs/output/edge/Poisson2D/TTD_VPINN/min_error_prediction.txt")
y_pred = y_pred.reshape(100, 100)
contour2 = axes[0,1].contourf(X, Y, y_pred, levels=100, cmap='jet')
axes[0,1].set_title('(b)', fontsize=18)
axes[0,1].set_xlabel('x', fontsize=18)
#axes[1].set_ylabel('y', fontsize=16)
axes[0,1].set_xticks([0, 0.5, 1])
axes[0,1].set_yticks([0, 0.5, 1])
axes[0,1].set_aspect('equal')

axes[0,1].set_xticklabels(['0', '0.5', '1'])
axes[0,1].set_yticklabels([])

plt.colorbar(contour2, ax=axes[0,1])  # Add colorbar to ax2

# Plot the error on ax3
error = np.loadtxt("/home/jovita/Projects/EdgeVPINNs/output/edge/Poisson2D/TTD_VPINN/min_error_error.txt")
error = error.reshape(100, 100)
contour3 = axes[0,2].contourf(X, Y, error, levels=100, cmap='jet')
axes[0,2].set_xlabel('x', fontsize=18)
#axes[2].set_ylabel('y', fontsize=16)
axes[0,2].set_title('(c)', fontsize=18)
axes[0,2].set_xticks([0, 0.5, 1])
axes[0,2].set_yticks([0, 0.5, 1])
axes[0,2].set_aspect('equal')
axes[0,2].set_yticklabels([])


axes[0,2].set_xticklabels(['0', '0.5', '1'])
plt.colorbar(contour3, ax=axes[0,2])  # Add colorbar to ax3

# Adjust layout for better spacing
rank = ttd['Rank']
l2_error = ttd['L2 Error - FastVPINNs']

# Define min and max L2 error
l2_error_min = min(l2_error)
l2_error_max = max(l2_error)

yticks = np.logspace(np.floor(np.log10(l2_error_min)), np.ceil(np.log10(l2_error_max)), num=5)
ytick_labels = [f"{y:.2e}" for y in yticks]  # Using scientific notation for readability
axes[1, 0].set_yticklabels(ytick_labels)# Create y-ticks (logarithmic scale) from min to max L2 error


# Plot the data on axes[1, 0]
axes[1, 0].plot(rank, l2_error, marker='o', color='darkblue')
axes[1, 0].axhline(y=1.931e-03, color='r', linestyle='--', label='Baseline L2 Error')

# Set y-axis to log scale, and add custom y-ticks and their corresponding labels
#axes[1, 0].set_yscale('log')
axes[1, 0].set_xticks(rank)
axes[1, 0].set_yticks(yticks)
#axes[1, 0].set_yticklabels(ytick_labels)

# Set labels and grid
axes[1, 0].set_xlabel('Rank', fontsize=18)
axes[1, 0].set_ylabel('L2 Error', fontsize=18)
axes[1,0].set_title('(d)', fontsize=18)

axes[1, 0].grid(True, which='both', linestyle='--', linewidth=0.7)

# Add legend to the subplot
axes[1, 0].legend()




# predicted solution - ttd rank 9




pred_ttd = np.loadtxt("/home/jovita/Projects/EdgeVPINNs/output/ttd_neural/7/min_error_prediction.txt")
pred_ttd = pred_ttd.reshape(100, 100)
contour4 = axes[1,1].contourf(X, Y, pred_ttd, levels=100, cmap='jet')
axes[1,1].set_xlabel('x', fontsize=18)
axes[1,1].set_ylabel('y', fontsize=18)
axes[1,1].set_title('(e)', fontsize=18)
axes[1,1].set_xticks([0, 0.5, 1])
axes[1,1].set_yticks([0, 0.5, 1])
axes[1,1].set_aspect('equal')
axes[1,1].set_yticklabels([])


axes[1,1].set_xticklabels(['0', '0.5', '1'])
plt.colorbar(contour4, ax=axes[1,1]) 


pred_ttd = np.loadtxt("/home/jovita/Projects/EdgeVPINNs/output/ttd_neural/7/min_error_error.txt")
pred_ttd = pred_ttd.reshape(100, 100)
contour5 = axes[1,2].contourf(X, Y, pred_ttd, levels=100, cmap='jet')
axes[1,2].set_xlabel('x', fontsize=18)
#axes[2].set_ylabel('y', fontsize=16)
axes[1,2].set_title('(f)', fontsize=18)
axes[1,2].set_xticks([0, 0.5, 1])
axes[1,2].set_yticks([0, 0.5, 1])
axes[1,2].set_aspect('equal')
axes[1,2].set_yticklabels([])


axes[1,2].set_xticklabels(['0', '0.5', '1'])
plt.colorbar(contour5, ax=axes[1,2]) 

plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=1.0)
plt.show()

```
